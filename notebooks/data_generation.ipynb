```python
# HR Dataset Generation using Faker
# This notebook generates a synthetic HR dataset used for attrition analytics.
```

```python
#import libraries
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
fake = Faker()
```

```python
num_employees = 35000
```

```python
locations = [
    "Chennai",
    "Bangalore",
    "Hyderabad",
    "Pune",
    "Mumbai"
]
```

```python
departments = [
    "Sales",
    "Engineering",
    "Human Resources",
    "Finance",
    "Marketing",
    "Operations"
]
```

```python
roles = [
    "Associate",
    "Senior Associate",
    "Lead",
    "Manager"
]
```

```python
salary_bands = ["Low","Medium","High"]
```

```python
data = []
```

```python
for i in range(1, num_employees + 1):

    first = fake.first_name()
    last = fake.last_name()

    email = (first + last + str(random.randint(1,999)) + "@gmail.com").lower()

    department = random.choice(departments)
    role = random.choice(roles)

    age = random.randint(22,55)

    experience = random.randint(0,15)

    tenure = round(np.random.uniform(0,10),2)

    satisfaction = round(np.random.normal(70,10),2)

    overtime = round(np.random.normal(10,5),2)

    attrition_prob = 0.15

    if satisfaction < 60:
        attrition_prob += 0.20

    if overtime > 15:
        attrition_prob += 0.15

    attrition = 1 if random.random() < attrition_prob else 0

    data.append([
        i,
        first,
        last,
        email,
        random.choice(["Male","Female"]),
        age,
        random.choice(locations),
        department,
        role,
        experience,
        tenure,
        random.choice(salary_bands),
        random.randint(1000,8000),
        round(np.random.uniform(1,5),2),
        random.randint(10,100),
        overtime,
        random.randint(1,10),
        satisfaction,
        round(np.random.uniform(1,5),2),
        attrition
    ])
```

```python
columns = [
"employee_id",
"first_name",
"last_name",
"email",
"gender",
"age",
"location",
"department",
"role",
"experience_years",
"tenure_years",
"salary_band",
"bonus",
"performance_score",
"training_hours",
"overtime_hours",
"projects_assigned",
"satisfaction_score",
"work_life_balance",
"attrition"
]
```

```python
df = pd.DataFrame(data, columns=columns)
```

```python
df["employee_id"] = range(1001, 1001 + len(df))
```

```python
df.loc[df.sample(frac=0.02).index,"training_hours"] = np.nan

df.loc[df.sample(frac=0.02).index,"satisfaction_score"] = np.nan
```

```python
df.loc[df.sample(frac=0.01).index,"overtime_hours"] = 35
```
